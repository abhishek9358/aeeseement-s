```python
import pandas as pd

sales_data = pd.read_csv(r'C:\Users\abhis\Desktop\Aspire_assessment\pandas_assessment\sales_data.csv')

sales_data.dropna(inplace=True)
pd.DataFrame(sales_data)
```

```python
grouped_sales = sales_data.groupby(['product_category', 'customer_segment']).agg(
    Total_Sales=('sales', 'sum'),
    Average_Discount=('discount', 'mean')
).reset_index()

grouped_sales.head()
```

```python
sales_data.dropna(inplace=True)

product_details = pd.DataFrame({
    'product_id': [1, 2, 3, 4, 5],  
    'Product_Name': ['Suger', 'Oil', 'Cloths', 'Soap', 'Powder'],  
    'Category': ['Grossery', 'Grossery', 'Fancy', 'Fancy', 'Fancy']  
})

merged_data = pd.merge(sales_data, product_details, on='product_id', how='left')

merged_data.head()
```

```python
sales_data['date'] = pd.to_datetime(sales_data['date'])

monthly_data = sales_data.resample('ME', on='date').mean()

monthly_data.head()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns


# Task 5: Additional exploratory data analysis or data manipulation tasks


plt.figure(figsize=(10, 6))
plt.plot(sales_data['Date'], sales_data['Sales'])
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
sns.histplot(sales_data['Sales'], bins=20, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()


correlation_matrix = sales_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
